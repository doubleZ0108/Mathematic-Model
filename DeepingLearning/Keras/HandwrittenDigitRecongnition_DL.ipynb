```python
import numpy as np
from PIL import Image
from HandwrittenDigitRecongnition import ImageDigit
```

```python
imageToDigit = ImageDigit('handwritting.png')
imageToDigit.histShow()
threshold = 143
imageToDigit.convert_to_bw(threshold)
imageToDigit.split()
imageToDigit.standardizing_to_32_32()
X, y = imageToDigit.featureExtract()
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=.1)  
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=100,activation='relu',input_dim=256))
model.add(Dense(units=10,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
model.fit(Xtrain, ytrain, epochs=100, batch_size=5)
classes = model.predict(Xtest, batch_size=5)
predict = np.argmax(classes, axis=1)
ytrue = np.argmax(ytest, axis=1)
err = ytrue - predict
print(err)
```
