```python
import numpy as np
X = np.loadtxt('wheat_train_PCA_X.txt')
y = np.loadtxt('wheat_train_PCA_Y.txt')
y[y!=1] = 0
from keras.utils import to_categorical
y = to_categorical(y)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=.1) 
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=50, activation='relu', input_dim=24)) #输入层=24，隐层=50
model.add(Dense(units=2, activation='softmax' ))  # ，输出2节点
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
model.fit(Xtrain, ytrain, epochs=500, batch_size=5)
#loss_and_metrics = model.evaluate(X, y, batch_size=10)
classes = model.predict(Xtest, batch_size=5)
predict=np.argmax(classes,axis=1)
ytrue=np.argmax(ytest,axis=1)
err=ytrue-predict
print('误差：',err)
```
